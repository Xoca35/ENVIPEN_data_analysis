```
def plot_parametros(x,hue,title,xlabel,n):
  #fig= sns.countplot(x="SEXO",data=df_delitos)
  ax= sns.countplot(x=x,data=df_delitos, hue=hue,edgecolor=(0,0,0),linewidth=2)
  ax.set_title(title)
  ax.set_xlabel(xlabel)
  ncount = len(df_delitos)
  # Make twin axis
  ax2=ax.twinx()

  # Switch so count axis is on right, frequency on left
  ax2.yaxis.tick_left()
  ax.yaxis.tick_right()
  ax.set_ylabel("Total")

  # Also switch the labels over
  ax.yaxis.set_label_position('right')
  ax2.yaxis.set_label_position('left')

  ax2.set_ylabel('Porcentaje [%]')

  for p in ax.patches:
      x=p.get_bbox().get_points()[:,0]
      y=p.get_bbox().get_points()[1,1]
      ax.annotate('{:.1f}%'.format(100.*y/ncount), (x.mean(), y), 
              ha='center', va='bottom') # set the alignment of the text

  # Use a LinearLocator to ensure the correct number of ticks
  ax.yaxis.set_major_locator(ticker.LinearLocator(11))

  # Fix the frequency range to 0-100
  ax2.set_ylim(0,100/n)
  ax.set_ylim(0,ncount/n)

  # And use a MultipleLocator to ensure a tick spacing of 10
  ax2.yaxis.set_major_locator(ticker.MultipleLocator(10))
```

```
plot_parametros("BP1_20",hue="SEXO",title="Porcentaje delitos denunciados",xlabel="Delitos denunciados",n=1.5)
```

```
plot_parametros("SEXO",hue=None,title="Porcentaje delitos denunciados",xlabel="Delitos denunciados",n=1.5)
```
